```python
import cv2
import glob
import os
```

```python
classes = {"0": "yu", "1": "shu", "2": "ryo"}
```

```python
def makedir(dir):
  if not os.path.exists(dir):
    os.makedirs(dir)

for key in classes:
  makedir(os.path.join('data', classes[key]))
```

```python
def trim_from_a_image(jpg_path):
  img = cv2.imread(jpg_path)
  txt_path = os.path.splitext(jpg_path)[0] + '.txt'
  count = 0
  with open(txt_path) as f:
    s = f.readlines()
    print(s)
    for line in s:
      split = line.split()
      if split:
        center_x_float = float(split[1])
        center_y_float = float(split[2])
        width_float    = float(split[3])
        height_float   = float(split[4])
    
        print(center_x_float)  
        print(center_y_float)
        print(width_float)  
        print(height_float)
        print(img.shape)
        center_x_int = round(img.shape[1] * center_x_float)
        center_y_int = round(img.shape[0] * center_y_float)
        width_int    = round(img.shape[1] * width_float)
        height_int   = round(img.shape[0] * height_float)
        kiri_left = center_x_int - width_int // 2
        kiri_right = kiri_left + width_int
        kiri_top = center_y_int - height_int // 2
        kiri_bottom = kiri_top + height_int
        kiri_img = img[kiri_top:kiri_bottom, kiri_left:kiri_right]
        cv2.imwrite(os.path.join('data',classes[split[0]],os.path.splitext(os.path.basename(jpg_path))[0] + "_" + str(count) + '.jpg'), kiri_img)
        count += 1
```

```python
for jpg_path in sorted(glob.glob("data/data3class_yolo/*.jpg")):
  trim_from_a_image(jpg_path)
```
